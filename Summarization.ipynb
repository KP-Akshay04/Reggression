```python
article_text = "This is the text you want to clean. It may contain spaces,  punctuations, and numbers."
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize

# Cleaning the text by removing numbers and special characters
clean_text = re.sub('[^a-zA-Z]', ' ', article_text)
print(clean_text)
```

```python
article_text = article_text.lower()
print(article_text)
```

```python
nltk.download('punkt')

sentence_list = nltk.sent_tokenize(clean_text)
print(sentence_list)
```

```python
tokens = word_tokenize(article_text)
print(tokens)
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')

# Create a dictionary to store word frequencies, ignoring stopwords
word_frequencies = {}
for word in tokens:
    if word not in stopwords:
        if word not in word_frequencies:
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1

print(word_frequencies)
```

```python
# Import string module for removing punctuation and digits
import string
from nltk.probability import FreqDist

# Remove punctuation and digits from stopwords list
remove_these = set(stopwords + list(string.punctuation) + list(string.digits))

# Create frequency distribution of the words
word_freq = FreqDist(word_frequencies)
print(word_freq)
```

```python
# Plot the frequency distribution using word frequencies
word_freq.plot(10, title="freqplot")
```

```python
# Import Counter from collections for counting frequencies
from collections import Counter

# Count the word frequencies
dictionary_counter = Counter(word_frequencies)
print(dictionary_counter)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Generate a word cloud based on word frequencies
cloud = WordCloud(max_font_size=80, colormap='hsv').generate_from_frequencies(dictionary_counter)

# Plot the word cloud
plt.figure(figsize=(16,12))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
